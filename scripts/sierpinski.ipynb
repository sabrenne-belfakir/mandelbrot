```python
import matplotlib.pyplot as plt

vertices = [[-10, -10], [10, -10], [0, 10]]
depth = 7

def sierpinski(vertices, depth):
    if depth == 0:
        triangle = plt.Polygon(vertices, edgecolor='red')
        plt.gca().add_patch(triangle)    
    else:    
        midab = [(vertices[0][0] + vertices[1][0]) / 2, (vertices[0][1] + vertices[1][1]) / 2]
        midac = [(vertices[0][0] + vertices[2][0]) / 2, (vertices[0][1] + vertices[2][1]) / 2]
        midbc = [(vertices[1][0] + vertices[2][0]) / 2, (vertices[1][1] + vertices[2][1]) / 2]

        sierpinski([vertices[0], midab, midac], depth -1)
        sierpinski([midab, vertices[1], midbc], depth -1)
        sierpinski([midac, midbc, vertices[2]], depth -1)


plt.figure(figsize=(12, 6))
sierpinski(vertices, depth)

plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.grid()
plt.show()
```
